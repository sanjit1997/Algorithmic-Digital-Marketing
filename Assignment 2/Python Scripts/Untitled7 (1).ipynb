```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
import pandas as pd
import requests
import urllib
import json
```

```python
page = urllib.request.urlopen("https://www.trustpilot.com/review/www.hyatt.com")
page_html=page.read()
html_soup=BeautifulSoup(page_html,'html.parser')
abc=str(html_soup)
start='"headline":'
end=',"reviewRating"'
answer=abc[abc.find(start)+len(start):abc.rfind(end)]
answer='"headline":'+answer
list1=answer.split(",")
print(list1)
import csv
list2=[]




```

```python
for i in list1:
  if re.search("reviewRating",i):
    continue
  if re.search("bestRating",i):
    continue
  if re.search("worstRating",i):
    continue
  if re.search("publisher",i):
    continue
  if re.search("name",i):
    continue
  if re.search("sameAs",i):
    continue
  if re.search("inLanguage",i):
    continue
  if re.search("@type",i):
    continue
  if re.search("itemReviewed",i):
    continue
  if re.search("image",i):
    continue
  if re.search("name",i):
    continue
  if re.search("url",i):
    continue
  if re.search("datePublished",i):
    continue
  else:
    txt = re.sub('[},"]', ' ', i)
    if re.search("headline",txt):
      list2.append(txt)
    if re.search("reviewBody",txt):
      list2.append(txt)
    if re.search("ratingValue",txt):
      list2.append(txt)
    if not re.search("headline",txt) and re.search("reviewBody",txt) and re.search("ratingValue",txt):
      txt1=list2[-1]
      list2.pop(-1)
      txt3=txt1+" "+txt
      print(txt3)
      list2.append(txt3)


    
```

```python
textfile = open("a_file2_new.txt", "w")
for element in list2:
    textfile.write(element + "\n")
textfile.close()
```
